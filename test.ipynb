```python
import torch 
import torchvision
import torchvision.transforms as transforms
from torchvision.io import read_image
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
train_dataset_path = './pothole600/training/'
validation_dataset_path = './pothole600/validation/'
test_dataset_path = './pothole600/testing/'

transformsCustom = transforms.Compose([
    transforms.ToTensor()
    ])

# train_dataset = torchvision.datasets.ImageFolder(root=train_dataset_path,transform=transformsCustom)
train_dataset = torchvision.datasets.ImageFolder(root=train_dataset_path,transform=transformsCustom)
validation_dataset = torchvision.datasets.ImageFolder(root=validation_dataset_path)
test_dataset = torchvision.datasets.ImageFolder(root=test_dataset_path)

train_loader = DataLoader(train_dataset, batch_size=6,shuffle=True)

def show_images(data_loader):
    batch = next(iter(data_loader))
    images, labels = batch

    grid = torchvision.utils.make_grid(images,nrow=3)
    plt.figure(figsize=(11,11))
    plt.imshow(np.transpose(grid,(1,2,0)))
    print('labels: ',labels)

show_images(train_loader)
```
